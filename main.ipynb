```python
from huggingface_hub import hf_hub_download
import pandas as pd

# Parámetros del dataset
repo_id = "sabareesh88/FNSPID_nasdaq"
repo_type = "dataset"
num_files = 48  # Total de archivos train según la estructura del dataset

# Lista para almacenar los DataFrames parciales
dfs = []

# Descargar y leer cada archivo
for i in range(num_files):
    file_name = f"data/train-{i:05d}-of-00048.parquet"
    print(f"Descargando y leyendo: {file_name}")

    # Descargar archivo real desde Hugging Face Hub
    file_path = hf_hub_download(repo_id=repo_id, filename=file_name, repo_type=repo_type)

    # Leer con pandas
    df = pd.read_parquet(file_path)
    dfs.append(df)

# Unir todos los DataFrames en uno solo
full_df = pd.concat(dfs, ignore_index=True)
print("✅ Dataset completo cargado con shape:", full_df.shape)

# Ver las primeras filas
full_df.head()
```

```python
full_df.info()
```

```python
df_apple = full_df[full_df['Stock_symbol'] == 'GOOG']
df_apple.head()
```

```python
df_apple = full_df
```

```python
df_apple.info()
```

```python
df_apple['Date'] = pd.to_datetime(df_apple['Date'])
```

```python
min_date = df_apple['Date'].min()
max_date = df_apple['Date'].max()

print("📅 Fecha mínima:", min_date)
print("📅 Fecha máxima:", max_date)
```

```python
registros_por_dia = df_apple.groupby(df_apple['Date'].dt.date).size().reset_index(name='n_registros')
```

```python
registros_por_dia.head()
```

```python
registros_por_dia.sort_values(by='n_registros', ascending=True).head(10)  # Top 10 días con más registros
```

```python
registros_por_dia['n_registros'].mean()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(15, 4))
plt.plot(registros_por_dia['Date'], registros_por_dia['n_registros'])
plt.title("Número de registros por día")
plt.xlabel("Fecha")
plt.ylabel("Cantidad")
plt.grid(True)
plt.tight_layout()
plt.show()
```
